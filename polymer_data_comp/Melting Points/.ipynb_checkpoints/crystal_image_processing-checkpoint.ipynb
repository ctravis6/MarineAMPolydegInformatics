```python
from tensorflow.keras.callbacks import ReduceLROnPlateau,ModelCheckpoint
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix
from tensorflow import keras
import matplotlib.pyplot as plt
import tensorflow as tf
from PIL import Image
import numpy as np  
import itertools
import json
import os
import pandas as pd
from sklearn.model_selection import train_test_split
import cv2,random
import umap,pickle
import seaborn as sns
import time
%matplotlib inline 
```

```python
nclasses = 4
nchannels = 3 #RGB
imagesize=(300,300)
batchsize=32
epochs=20
```

```python
data_dir = '/gdrive/Shareddrives/uwds4mat_20a/Database/MARCO/'
model_dir = data_dir + 'model/'
model_path = model_dir+'MARCO_300.{epoch:02d}-{val_accuracy:.4f}.hdf5'
result_dir = data_dir+'result/'

df_all = pd.read_csv(data_dir+'R_smallset300.csv',index_col=None)
df_all['image_path'] = df_all.apply(lambda row:data_dir+row['image_path'],axis=1)
df_all.head()
```

```python
print(df_all['image_path'][0])
print(os.path.exists(df_all.iloc[0].image_path))
```

```python
df_all.groupby(['label_text']).count() #show samples for each classes
```

```python
#get the training dataset 
seed = 42

Rcrystal = df_all[df_all["label_id"]==1].sample(2000,random_state=seed).copy().reset_index(drop=True)
Rother = df_all[df_all["label_id"]==2].sample(2000,random_state=seed).copy().reset_index(drop=True)
Rpreci = df_all[df_all["label_id"]==3].sample(2000,random_state=seed).copy().reset_index(drop=True)
Rclear = df_all[df_all["label_id"]==0].sample(2000,random_state=seed).copy().reset_index(drop=True)
df = pd.concat([Rcrystal,Rother,Rpreci,Rclear],0,ignore_index=True,sort=False)
```

```python
Rcrystal.head()
```

```python
#get the testing dataset 
seed = 42
def getTestData(df_train,df_all,classid,nsample):
  R_test = pd.concat([df_train,df_all],0,ignore_index=True,sort=False).drop_duplicates(keep=False) #drop all repeat ones
  Res_test = R_test[R_test["label_id"]==classid].sample(nsample,random_state=seed).copy().reset_index(drop=True) #select from left ones
  return(Res_test)
Rclear_test = getTestData(Rclear,df_all,0,1000)
Rcrystal_test = getTestData(Rcrystal,df_all,1,1000)
Rother_test = getTestData(Rother,df_all,2,1000)
Rpreci_test = getTestData(Rpreci,df_all,3,1000)


df_test = pd.concat([Rcrystal_test,Rother_test,Rpreci_test,Rclear_test],0,ignore_index=True,sort=False)
```

```python
print(len(df_test))
df_test.head()
```

```python
df_train, df_val = train_test_split(df, test_size=0.2, stratify=df['label_id'],random_state=42)
print("Number of training and validation samples are respectively:", len(df_train), "and",len(df_val))
```

```python
df_train.head()
```

```python
# Rescale all images by 1./255 and apply image augmentation
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
                rotation_range=15,
                width_shift_range=0.1,
                height_shift_range=0.1,
                zoom_range=[0.9, 1.2],
                horizontal_flip=True,
                vertical_flip=True,
                fill_mode='reflect',
                rescale=1./255
                )

validation_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
                rescale=1./255
                )
```

```python
train_generator = train_datagen.flow_from_dataframe(
                df_train,
                x_col = 'image_path',  
                y_col ='label_text',
                target_size=imagesize,  
                batch_size=batchsize,
                shuffle = True,
                class_mode='categorical')

# Flow validation images in batches of 20 using test_datagen generator
validation_generator = validation_datagen.flow_from_dataframe(
                df_val,
                x_col = 'image_path',  
                y_col ='label_text',
                target_size=imagesize,  
                batch_size=batchsize,
                shuffle = False,
                class_mode='categorical')
```

```python
basemodel = tf.keras.applications.DenseNet121(weights='imagenet',include_top=False,input_shape=(imagesize[0],imagesize[1],nchannels))
basemodel.trainable = True
#Freeze the previous layer and train the last 5 layers
#for layers in basemodel.layers[:-5]:
#    layers.trainable = False
```

```python
#Build model    
model = tf.keras.Sequential()
model.add(basemodel)
model.add(tf.keras.layers.GlobalAveragePooling2D())  #Join the global average pooling layer
model.add(tf.keras.layers.Dense(512,activation='relu'))  #Add fully connected layer
model.add(tf.keras.layers.Dropout(rate=0.25))  #Add Dropout layer to prevent overfitting
model.add(tf.keras.layers.Dense(128,activation='relu'))  #Add fully connected layer
model.add(tf.keras.layers.Dropout(rate=0.25))  #Add Dropout layer to prevent overfitting
model.add(tf.keras.layers.Dense(nclasses,activation='softmax'))  #Add output layer (2 categories)
model.summary()   #Print each layer parameter information 
```

```python
model.save(model_dir+'marco300.h5') 
```

```python
#Compile model
initial_learning_rate = 0.0002
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=initial_learning_rate), #Use 'Adam' optimizer, the learning rate is 0.0001
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=False), #Cross entropy loss function
              metrics=["accuracy"]) #Evaluation function
```

```python
#Callback function 1: Learning rate decay
reduce_lr = ReduceLROnPlateau(
                                monitor='val_loss', #The value to be monitored
                                factor=0.7,  #Learning rate decays to 1/10 of the original
                                patience=3,  #When patience epochs are past and the model performance does not improve, the action of reducing the learning rate will be triggered
                                mode='auto', #When the monitoring value is val_acc, the mode should be max. When the monitoring value is val_loss, the mode should be min. In auto mode, the evaluation criterion is automatically inferred from the name of the monitored value
                                verbose=1 #If True, output a message for each update, default value: False
                             )
#Callback function 2: Save the optimal model
checkpoint = ModelCheckpoint(
                                filepath=model_path, # Path at which to save the model file.
                                monitor='val_accuracy', #Need to monitor the value
                                save_weights_only=True, #If set to True, only the model weight will be saved, otherwise the entire model (including model structure, configuration information, etc.) will be saved
                                save_best_only=True, #When set to True, the current model will only be saved when the monitoring value is improved
                                mode='auto', #When the monitoring value is val_acc, the mode should be max. When the monitoring value is val_loss, the mode should be min. In auto mode, the evaluation criterion is automatically inferred from the name of the monitored value
                                verbose=1 #If True, output a message for each update, default value: False
                            )
```

```python
#Start training
start_time = time.time()
history = model.fit(x=train_generator,   #Enter training set
                    steps_per_epoch=train_generator.n // batchsize, #The number of training steps included in an epoch
                    epochs=epochs, #Training model iterations
                    validation_data=validation_generator,  #Enter validation set
                    validation_steps=validation_generator.n // batchsize, #The number of validation steps included in an epoch
                    workers = 16,
                    callbacks=[checkpoint, reduce_lr] #Execute callback function
                    )
total_time = time.time() - start_time
print(total_time)
```

```python
#prediction
#load in model and weight 
weightfile = model_dir+'MARCO_300.20-0.9256.hdf5'
model = keras.models.load_model(model_dir+'marco300.h5')
model.load_weights(weightfile) 
#create a model to predict the feature space. 
feature_model = tf.keras.Model(inputs=model.input, outputs=model.get_layer('dense_4').output)
```

```python
test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
                rescale=1./255
                )
```

```python
test_generator = test_datagen.flow_from_dataframe(
                df_test,
                x_col = 'image_path', 
                y_co = None, 
                target_size=imagesize,  
                batch_size=batchsize,
                shuffle = False,
                class_mode=None)

pred=model.predict(
    test_generator,
    steps=test_generator.n//batchsize,
    workers=8,
    verbose=1)

featpred=feature_model.predict(
    test_generator,
    workers=8,
    steps=test_generator.n//batchsize,
    verbose=1)
print(pred.shape)
```

```python
marco_dict ={0:'Clear',
             1:'Crystals',
             2:'Other',
             3:'Precipitate'}
pred_idx=np.argmax(pred,axis=1) # get predict index. 
df_test['pred_index'] = pred_idx
df_test['pred_label'] = df_test['pred_index'].map(marco_dict.get) 
```

```python
#save features into the test dataframe
featcols = ['Feature{0:03d}'.format(ii) for ii in range(128)]
for ii in range(len(featcols)):
  df_test[featcols[ii]] = featpred[:,ii]
```
